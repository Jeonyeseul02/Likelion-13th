```python
#목표 : 교보문고 베스트셀러(100위) 데이터를 가져와 분류별 선호도 알아보기

## 1 .가져와야할 데이터 : 1) 베스트 종합 셀러 1~5page 2) 100개 순위의 책이 속한 분류
## 2. 1-2)번의 데이터를 len함수를 이용하여 분류별로 몇개가 랭크되어있는지를 파악한다.
## 3. 1-1)데이터를 엑셀 파일로 만든다 / 2번의 데이터를 그래프화(?) 시킨다.
## 3. 결과물을 ppt or github 작업화 한다.


from selenium import  webdriver
from bs4 import BeautifulSoup
import time
import pandas as pd

url = "http://www.kyobobook.co.kr/bestSellerNew/bestseller.laf"
driver = webdriver.Chrome('./chromedriver')

driver.get(url)
```

```python
# 1~-5page(100개)의 베스트셀러 정보를 가져온다
all_best = []
for num in range(1,21,1):
     xpath_url = '//*[@id="main_contents"]/ul/li[' +str(num)+ ']/div[2]/div[2]/a/strong'
     one_best = driver.find_element_by_xpath(xpath_url)
     all_best.append(one_best.text)

for btn in range(2,6,1):
    sel_btn = driver.find_element_by_xpath('//*[@id="main_contents"]/div[5]/div[1]/ul/li[' +str(btn)+ ']/a')
    sel_btn.click()
    time.sleep(3)
    for num in range(1,21,1):
        xpath_url = '//*[@id="main_contents"]/ul/li[' +str(num)+ ']/div[2]/div[2]/a/strong'
        one_best = driver.find_element_by_xpath(xpath_url)
        all_best.append(one_best.text)
all_best
```

```python
#가져온 정보가 제대로 된 지 len함수를 써서 확인한다.
print(len(all_best))
```

```python
# 다시 1페이지로 돌아간다.(이전 작업 : 5페이지)
sel_btn_1 = driver.find_element_by_xpath('//*[@id="main_contents"]/div[4]/div[1]/ul/li[1]/a')
sel_btn_1.click()
```

```python
## 1위 분류##  //*[@id="container"]/div[5]/div[1]/div[3]/ul/li[1]/a[1]
## 2위 분류## //*[@id="container"]/div[5]/div[1]/div[3]/ul/li[1]/a[1]

# 1~5page 분류별 정보를 가져온다. 
#(베스트셀러와 다르게 한 페이지에 20개의 정보를 담고 있는 것이 아니여서
#데이터(베스트셀러)를 하나하나 클릭하여 되돌아가기(back) 기능을 써서 1~5page 데이터를 가져옴
#시간이,,,, 5~8분정도 소요된다.,,,,,,,ㅠ___ㅠ

all_sector = []
for btn2 in range(1,21,1):
    sel_btn = driver.find_element_by_xpath('//*[@id="main_contents"]/ul/li[' +str(btn2)+ ']/div[2]/div[2]/a/strong')
    sel_btn.click()
    time.sleep(1)
    xpath = '//*[@id="container"]/div[5]/div[1]/div[3]/ul/li[1]/a[1]'
    sector = driver.find_element_by_xpath(xpath)
    all_sector.append(sector.text)
    driver.back()
    time.sleep(1)

for page in range(2,6,1):
    sel_page = driver.find_element_by_xpath('//*[@id="main_contents"]/div[5]/div[1]/ul/li[' +str(page)+ ']/a')
    sel_page.click()
    for btn2 in range(1,21,1):
        sel_btn = driver.find_element_by_xpath('//*[@id="main_contents"]/ul/li[' +str(btn2)+ ']/div[2]/div[2]/a/strong')
        sel_btn.click()
        time.sleep(1)
        xpath = '//*[@id="container"]/div[5]/div[1]/div[3]/ul/li[1]/a[1]'
        sector = driver.find_element_by_xpath(xpath)
        all_sector.append(sector.text)
        driver.back()
        time.sleep(1)
        
all_sector
```

```python
#가져온 정보가 맞는지 확인
print(len(all_sector))
```

```python
#가져온 정보를 딕셔너리로 묶어 csv/excel 파일화 시킨다.
dic_dict = {"베스트셀러":all_best,"분류":all_sector}
dat = pd.DataFrame(dic_dict)
dat.to_csv("교보문고_베스트셀러_100.csv", index=False)
dat.to_excel("교보문고_베스트셀러_100.xlsx")
```
