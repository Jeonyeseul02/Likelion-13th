```python
import pandas as pd
import folium
```

```python
dat = pd.read_csv("서울시 금천구 CCTV 설치 위치정보.csv",engine='python',encoding='EUCKR')
dat
```

```python
dat_tmp = dat.loc[ :100, ['명칭', '위도', '경도' ] ]
dat_tmp
```

```python
# 중앙 위치 찾기
dat_tmp.columns = ['name','latitude','longitude']
print(dat_tmp.latitude.mean(), dat_tmp.longitude.mean())
dat_tmp.head()
```

```python
des = dat_tmp['name']
dat_p1 = list(dat_tmp['latitude'])
dat_p2 = list(dat_tmp['longitude'])
```

```python
dat_xy = list(zip(dat_p1, dat_p2))
```

```python
m = folium.Map(location=[37.475977,126.906818 ], zoom_start=13)

for name,loc in zip(des, dat_xy):
  folium.Marker(location=loc, tooltip=name, icon=folium.Icon(color='purple', icon='info-sign')).add_to(m)
m.save("seoul_금천구_CCTV_100.html")
m
```
