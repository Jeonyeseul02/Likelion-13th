```python
import nltk
from konlpy.tag import Kkma
from konlpy.tag import Hannanum
```

```python
from wordcloud import WordCloud, STOPWORDS
from PIL import Image
```

```python
text = open("코로나_뉴스_1000.csv").read()
text
```

```python
print(STOPWORDS)
```

```python
from konlpy.tag import Okt   ### Okt
t = Okt()

#from konlpy.tag import Kkma  ### Okt
#t = Kkma()
doc_nouns = t.nouns(text)
print( len( doc_nouns ) )
doc_nouns
```

```python
ko = nltk.Text(doc_nouns, name="코로나뉴스_1000")
print(type(ko))
print(len(ko.tokens))
```

```python
### 단어들의 사용 횟수 확인 - 빈도 분석
type( ko.vocab() ), ko.vocab().most_common(20)
```

```python
most_fre = ko.vocab().most_common(50)
most_fre
```

```python
### 한글에서는 따로 불용어 사전이 없어, 따로 만들거나 또는 파일로 부터 불러올 수 있다.
stop_words = ['진자','명','확','차','제','등','더','속','로','세','률','선','것']

new_ko = [ ]
for one_word in ko:
  if one_word not in stop_words:
    new_ko.append(one_word)
```

```python
new_ko = nltk.Text(new_ko, name="코로나_뉴스_1000_2")
```

```python
from wordcloud import WordCloud, STOPWORDS

import numpy as np
from PIL import Image
```

```python
data = new_ko.vocab().most_common(1000)
```

```python
wc = WordCloud(background_color='white', 
               max_words=200,
               contour_width=3, 
               contour_color="steelblue", 
               font_path=path).generate_from_frequencies(dict(data))
```

```python
plt.figure(figsize=(15,8))
plt.imshow(wc)
plt.axis("off")
plt.show()
```
