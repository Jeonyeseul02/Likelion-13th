```python
from pykrx import stock
```

### 2020.01.20 - 2021.08.31 기간 kospi 투자자별 거래량 

```python
investors_all = stock.get_market_trading_value_by_date("20200120", "20210831", "KOSPI")
investors_all
```

### 2020.01.20 - 2021.08.31 기간 kospi 투자자별 매도량

```python
investors_sell = stock.get_market_trading_value_by_date("20200120", "20210831", "KOSPI", on='매도')
investors_sell
```

### 2020.01.20 - 2021.08.31 기간 kospi 투자자별 매도량

```python
investor_get = stock.get_market_trading_value_by_date("20200120", "20210831", "KOSPI", on='매수')
investor_get
```

### csv, excel 파일로 저장

```python
investors_all.to_csv("kospi_investors_all.csv", index=False)
investors_all.to_excel("kospi_investors_all.xlsx", index=False)
```

```python
investors_sell.to_csv("kospi_investors_sell.csv", index=False)
investors_sell.to_excel("kospi_investors_sell.xlsx", index=False)
```

```python
investor_get.to_csv("kospi_investors_get.csv", index=False)
investor_get.to_excel("kospi_investors_get.xlsx", index=False)
```
