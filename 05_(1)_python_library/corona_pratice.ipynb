```python
from selenium import webdriver
from bs4 import BeautifulSoup

driver = webdriver.Chrome('./chromedriver')

url = 'https://www.bloomberg.com/graphics/covid-vaccine-tracker-global-distribution/'
driver.get(url)
```

```python
# //*[@id="dvz-table-usa-vaccination"]/div[2]/div[2]/button
sel_more2 = driver.find_element_by_xpath('//*[@id="dvz-table-global-vaccination"]/div[2]/div[2]/button')
sel_more2.click()
```

```python
all_data = []

for i in range(1, 7, 1):
    data_col = []
    xpath = '//*[@id="dvz-table-global-vaccination"]/div[2]/div[1]/table/tbody/tr/td[%s]' % str(i) 
    sel_data = driver.find_elements_by_xpath(xpath)
    
    for dat in sel_data:
        #print(dat)
        data_col.append(dat.text)
    print(data_col)
    all_data.append(data_col)
```

```python
import pandas as pd
```

```python
pd.set_option("display.max_rows", 40)

dat_dict = {'국가':all_data[0],
            '백신접종수':all_data[1],
            '백신보유율':all_data[2],
            '1차접종':all_data[3],
            '2차접종':all_data[4],
            '일별접종수':all_data[5]
           }

dat_df = pd.DataFrame(dat_dict)
dat_df
```

```python
import pandas as pd

# excel 데이터 셋도 pd.read_excel로 가져올 수 있음.
corona = pd.read_csv("./data/2021-09-28_corona.csv")
vaccine = pd.read_csv("./data/20210929_18_vaccine_bloomberg.csv")
country_code = pd.read_excel("./data/country.xlsx")

corona.shape, vaccine.shape, country_code.shape
```

```python
#코로나 csv 파일에 국가명에 약자가 있는걸 없앰.
corona['국가'].str.extract('([ㄱ-ㅣ가-힣]+)')
```

```python
#약자를 뺀값을 한글표기에 넣음
corona['한글표기'] = corona['국가'].str.extract('([ㄱ-ㅣ가-힣]+)', expand=False)
corona.columns
```

```python
col = ['국가', '한글표기', '치명(%)', '완치(%)', '발생률', 
       '인구수', '치료중합계','치료중1일','확진자합계', '확진자1일', '사망자합계', '사망자1일', '완치합계', '완치1일']

new_corona = corona[col].copy()
new_corona
```

```python
country_code.head()
```

```python
df_corona = new_corona.merge(country_code, left_on='한글표기', 
                             right_on='국문(코로나보드)')

df_corona.head()
```

```python
df_corona.drop(['국가', '나라', '국문(코로나보드)', '나라','비고_뜻'], axis=1, inplace=True)
df_corona.head()
```

```python
df_corona_all = df_corona.merge(vaccine, left_on='영문(블룸버그)', 
                             right_on='국가')

df_corona_all.head()
```

```python
df_corona_all.drop(['국가'], axis=1, inplace=True)
df_corona_all
```
